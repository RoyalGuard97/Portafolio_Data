Dada la gran cantidad de películas y series disponibles en la plataforma, es una oportunidad perfecta para mejorar sus habilidades de análisis de datos exploratorios y sumergirse en la industria del entretenimiento. Nuestro amigo también ha estado repasando sus habilidades con Python y ha probado por primera vez un archivo CSV que contiene datos de Netflix. Creen que la duración media de las películas ha ido disminuyendo. Usando la investigación inicial de tus amigos, profundizarás en los datos de Netflix para ver si puedes determinar si la duración de las películas en realidad se está acortando y explicar algunos de los factores que contribuyen, si los hay.

## Instrucciones
Tu amigo sospecha que las películas son cada vez más cortas y ha encontrado algunas pruebas iniciales de ello. Una vez que haya alcanzado su punto máximo de interés, realizará un análisis exploratorio de datos en los datos de netflix_data.csv para comprender qué puede estar contribuyendo a que las películas se acorten con el tiempo. Su análisis seguirá estos pasos:

1. Cargue el archivo CSV y guárdelo como netflix_df.
2. Filtre los datos para eliminar los programas de TV y guárdelos como netflix_subset.
3. Investigue los datos de la película de Netflix, conservando solo las columnas "título", "país", "género", "año_de_lanzamiento", "duración" y guardándolos en un nuevo DataFrame llamado netflix_movies.
4. Filtre netflix_movies para encontrar las películas de menos de 60 minutos, guardando el DataFrame resultante como short_movies; inspeccionar el resultado para encontrar posibles factores contribuyentes.
5. Usando un bucle for y sentencias if/elif, recorra las filas de netflix_movies y asigne colores de su elección a cuatro grupos de género ("Niños", "Documentales", "Stand-Up" y "Otros" para todo lo demás ). Guarde los resultados en una lista de colores.
6. Inicialice un objeto de figura llamado fig y cree un diagrama de dispersión para la duración de la película por año de lanzamiento usando la lista de colores para colorear los puntos y usando las etiquetas "Año de lanzamiento" para el eje x, "Duración (min)" para el y -eje y el título "Duración de la película por año de lanzamiento".
7. Después de inspeccionar la trama, responda la pregunta "¿Estamos seguros de que las películas se están acortando?" asignando "sí", "no" o "tal vez" a la respuesta variable.
8. Haga clic en el botón "Enviar proyecto" para comprobar su solución.
