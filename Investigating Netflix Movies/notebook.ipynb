## Cargue el archivo CSV y guárdelo como netflix_df.

```python
# Importamos las librerías
import pandas as pd
import warnings
import seaborn as sns
import matplotlib.pyplot as plt
warnings.filterwarnings("ignore")
```

```python
# Leemos el CSV
netflix_df = pd.read_csv("netflix_data.csv")
netflix_df
```

```python
# Vemos el nombre de las columnas
netflix_df.keys().to_list()
```

```python
# Otra manera de ver las columnas
netflix_df.columns.to_list()
```

```python
# Vemos que categorias hay en nuestra dataset
netflix_df["type"].unique()
```

## Filtre los datos para eliminar los programas de TV y guárdelos como netflix_subset.

```python
# Seleccionamos los registros que sean de tipo TV Show y lo guardamos en un DF
netflix_subset = netflix_df[netflix_df["type"].isin(["Movie"])]

# Otras maneras
# netflix_subset = netflix_df[~netflix_df["type"].isin(["TV Show"])]
# netflix_subset = netflix_df.drop(netflix_df[netflix_df["type"] == "TV Show"].index)
```

```python
# Lo visualizamos
netflix_subset
```

```python
# Comprobamos que solo queda la categorías Movies
netflix_subset["type"].unique().tolist()
```

## Investigue los datos de la película de Netflix, conservando solo las columnas "título", "país", "género", "año_de_lanzamiento", "duración" y guardándolos en un nuevo DataFrame llamado netflix_movies.

```python
# Visualizamos el dataset
netflix_subset
```

```python
# Extraemos las columnas
columnas_conservadas = ["title", "country", "genre", "release_year", "duration"]

# Lo guardamos en un DF
netflix_movies = netflix_subset[columnas_conservadas]
```

```python
# Visualizamos
netflix_movies
```

## Usando un bucle for y sentencias if/elif, recorra las filas de netflix_movies y asigne colores de su elección a cuatro grupos de género ("Niños", "Documentales", "Stand-Up" y "Otros" para todo lo demás ). Guarde los resultados en una lista de colores.

```python
# Vemos que elementos únicos tiene la columan Género
netflix_movies["genre"].unique()
```

```python
# Creamos la lista
colores = []

# Recorremos
for genero in netflix_movies["genre"].values:
    
    # Asignamos los colores
    if genero == "Children":
        # Agregamos a la lista
        colores.append("Red")
    elif genero == "Documentaries":
        colores.append("Blue")
    elif genero == "Stand-Up":
        colores.append("Yellow")
    else:
        colores.append("Black")

print(colores)
```

```python
# Otra manera usando el método iterrows
# Creamos la lista que almacenará los colores
colores = []

# Recorremos las filas del DF
for index, row in netflix_movies.iterrows():
    
    # Establecemos los generos
    genero = row["genre"]

    # Asignamos los colores
    if genero == "Children":
        # Agregamos a la lista
        colores.append("Red")
    elif genero == "Documentaries":
        colores.append("Blue")
    elif genero == "Stand-Up":
        colores.append("Yellow")
    else:
        colores.append("Black")
        
# Visualizamos
print(colores)    
```

```python
# Por lista de Compresión
colores = ["RED" if genero == "Children" else "BLUE" if genero == "Documentaries" else "YELLOW" if genero == "Stand-Up" else "GREEN" for genero in netflix_movies["genre"].values]

# Establecemos a minúsculas
colores = [color.lower() for color in colores]

# Visualizamos
print(colores)
```

```python
# Manera tradicional con enumerate estableciendo en mayúsculas
for index, color in enumerate(colores):
    colores[index] = color.upper()

# Visualizamos
print(colores)
```

## Inicialice un objeto de figura llamado fig y cree un diagrama de dispersión para la duración de la película por año de lanzamiento usando la lista de colores para colorear los puntos y usando las etiquetas "Año de lanzamiento" para el eje x, "Duración (min)" para el y -eje y el título "Duración de la película por año de lanzamiento".

```python
# Establecemos los ejes
x = netflix_movies["release_year"]
y = netflix_movies["duration"]

# Inicializamos la figura
fig = plt.figure()

# Creamos el gráfico de dispersión
plt.scatter(x, y, c=colores)

# Agregamos elementos
plt.xlabel("Año de Lanzamiento")
plt.ylabel("Duración en min")
plt.title("Duración de la Película por Año de Lanzamiento")

# Mostramos
plt.show()
```

```python
# Visualizamos por cada uno de los Géneros
sns.lmplot(data=netflix_movies, x="release_year", y="duration", hue='genre', col='genre',col_wrap=5)
```
